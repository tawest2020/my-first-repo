### Question 5

```r
getwd()
p_names <- read.csv("patient_names.csv")
p_props <- read.csv("patient_properties.csv")
p_names <- p_names %>%
  mutate(
    BIRTHDATE = as.Date(BIRTHDATE, format = "%m/%d/%y"),
    DEATHDATE = as.Date(DEATHDATE, format = "%m/%d/%y")
  )

head(p_names, 10)
head(p_props, 10)
```

### Question 6

```r
library(dplyr)
library(tidyr)
p_props_wide <- p_props %>%
pivot_wider(
    names_from = property,
    values_from = value
    )
head(p_props_wide, 10)
```

### Question 7

```r
p_join <- p_names %>%
left_join(p_props_wide, by = "ID")
head(p_join, 10)
```

### Question 8

```r
p_join$FIRST <- gsub("[0-9]+$", "", p_join$FIRST)
p_join$LAST <- gsub("[0-9]+$", "", p_join$LAST)
head(p_join, 10)
```

### Question 9

```r
p_cats <- setdiff(names(p_join), c("ID", "FIRST", "LAST"))
for (var in p_cats) {
    if(is.character(p_join[[var]]) || is.factor(p_join[[var]]) || inherits(p_join[[var]], "Date")) {
        counts <- p_join %>%
        count(.data[[var]]) %>%
        arrange(desc(n))
        cat("\n###", var, "\n")
        mdpre(counts)
        }
    }
```

### Question 10

Items to address
1. BIRTHDATE: R reading anything 1968 and prior as 2068 and prior
2. BIRTHDATE: The patient_names dataset reading anything from the 1920s as 2020s
3. MARITAL: Fine, male, and NA (1)
4. RACE: asiann (1)
5. ETHNICITY: nonhisapni (2) and hispani (1)
6. GENDER: Female, Male, and female (1)

```r
#1&2
p_join <- p_join %>%
mutate(
    BIRTHDATE = if_else(year(BIRTHDATE) >= 2000 & year(BIRTHDATE) <= 2099,
                        BIRTHDATE %m-% years(100),
                        BIRTHDATE)
  )
head(p_join, 10)
```

```r
#3-6
p_join <- p_join %>%
  mutate(
    MARITAL = if_else(MARITAL %in% c("Fine", "male"), NA_character_, MARITAL),
      RACE = if_else(RACE == "asiann", "asian", RACE),
      ETHNICITY = if_else(ETHNICITY == "nonhispani", "nonhispanic", ETHNICITY),
      ETHNICITY = if_else(ETHNICITY == "hispani", "hispanic", ETHNICITY),
      GENDER = if_else(GENDER == "Male", "M", GENDER),
      GENDER = if_else(GENDER == "Female", "F", GENDER),
      GENDER = if_else(GENDER == "female", "F", GENDER)
        )
p_cats_clean <- setdiff(names(p_join), c("ID", "FIRST", "LAST"))
for (var in p_cats) {
    if(is.character(p_join[[var]]) || is.factor(p_join[[var]]) || inherits(p_join[[var]], "Date")) {
        counts <- p_join %>%
        count(.data[[var]]) %>%
        arrange(desc(n))
        cat("\n###", var, "\n")
        mdpre(counts)
        }
    }
```

### Question 11

```r
p_join <- p_join %>%
  mutate(
    AGE = if_else(
      is.na(DEATHDATE),
      as.numeric(difftime(Sys.Date(), BIRTHDATE, units = "days")) / 365.25,
      as.numeric(difftime(DEATHDATE, BIRTHDATE, units = "days")) / 365.25,
    ),
      GENDER = recode(GENDER, "F" = "Female", "M" = "Male")
  )
library(ggplot2)
ggplot(p_join, aes(x = AGE, fill = GENDER)) +
  geom_histogram(binwidth = 5, position = "dodge", color = "black") +
  labs(
    title = "Distribution of Patient Age by Gender",
    x = "Age (years)",
    y = "Count",
    fill = "Gender"
  )
```
