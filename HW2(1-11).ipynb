```r
# Question 6: Differences in the way R processes...
x <- 0;
while (x <= 3) {
x <- x + 1;
}
x
## [1] 4
"x begins at 0 and continues to build off the previous function and completes its final run when x starts at 3 (automatic)"
for(y in c(1,2,3)) {
y <- y + 1;
}
y
## [1] 4
1
"Each function, y is assigned a different number that adds one, with the final function using the final number listed (manual)"
```

```r
# Question 9: Which counter function works... and why...
counter1 <- function(start, step){
val <- start;
function(){
old_val <- val;
val <- val + step;
old_val;
}
}
counter_from_1 <- counter1(1,1);
counter_from_1()
## [1] 1
counter_from_1()
## [1] 1
"counter1 fails to build off previous functions due to using '<- val + step' assignment, each cell will return 1"
"<- creates a binding in the current environment"
counter2 <- function(start, step){
val <- start;
function(){
old_val <- val;
val <<- val + step;
old_val;
}
}
counter_from_1 <- counter2(1,1);
counter_from_1()
## [1] 1
counter_from_1()
## [1] 2
"counter 2 builds off of previous functions thanks to the use of '<<- val + step' assignment, each cell will increase by 1"
"<<- updates/modifies the nearest enlosing binding"
```

```r
# Question 11: Fibonacci sequence
fib1 <- function() {
    a <- 0
    b <- 1
    function() {
        result <- a
        temp <- a + b
        a <<- b
        b <<- temp
        result
        }
    }
fib2 <- fib1()
fib3 <- numeric(7)
for (i in 1:7) {
    fib3[i] <- fib2()
    }
print(fib3)
```
